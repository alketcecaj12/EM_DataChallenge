### Time series decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# replace Nans with backward values or forward values 
series = data["carbon_intensity_avg"].bfill().ffill()
```

```python
# check for nans
series.isna().sum()
```

```python
data_decom = seasonal_decompose(series.values, model = 'additive', extrapolate_trend='freq', period=int(len(series)/24))
```

```python
with plt.rc_context(): 
    plt.rc("figure", figsize = (20, 10))
    data_decom.plot()
    plt.show()
```

### Insights from graphics above
- there is quite a lot of noise (random walk) in the data: the residual component
- there is a little bit of seasonality and a strong trend

```python
new_cols = []
for i in data.columns: 
    if i.startswith("latest_forecasted_"):
        new_cols.append(i)

new_cols.append("carbon_intensity_avg")
new_cols
```

```python
new_data = data.copy()
new_data = new_data[new_cols]

for i in new_data.columns: 
    new_data[i] = new_data[i].bfill().ffill()

new_data.drop("latest_forecasted_power_net_import_SE_avg", axis = 1, inplace = True)
new_data.isna().sum()
```

## How much are the features correlated to the variable **carbon_intensity_avg**

```python
import seaborn as sns
correlation_matrix = new_data.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()
```

## Insights from graphics above: 

- carbon_intensity_avg correlates well with different variables both positively and negatively. 

- if those correlations are not spurious then multivariate forecasting could be a good approach.
